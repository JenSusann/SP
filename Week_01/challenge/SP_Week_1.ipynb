```python
import numpy as np
random_numbers = np.random.random(100)
Q1 = np.percentile(random_numbers, 25)
Q2 = np.percentile(random_numbers, 50)
Q3 = np.percentile(random_numbers, 75)

print("Q1: ", Q1)
print("Q2: ", Q2)
print("Q3: ", Q3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

random_numbers = np.random.randn(1000)
plt.hist(random_numbers, bins=30, edgecolor='pink')
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
random_numbers = np.random.rand(50)
plt.boxplot(random_numbers)
plt.title('Boxplot of Random Numbers')
plt.ylabel('Value')

plt.show()
```
